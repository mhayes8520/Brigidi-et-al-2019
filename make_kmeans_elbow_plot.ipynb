```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import sys
import matplotlib.pyplot as plt
plt.switch_backend('agg')
import seaborn as sns
import matplotlib
from collections import Counter

print(matplotlib.__version__)

"""
Cluster Arnt/Arnt2 ChIP data using KMeans
"""

file = '/Users/michaelghayes/Documents/Svenner Shared/People/Michael/Experiments/180507 - Stefano Rat ChIP (180501) analysis/txt files/fileNameMatrix_arnt1-2_all-Npas4-peaks_3xGreaterPTX-vs-TTX.txt'
df = pd.read_csv(file, sep='\t')
#df.set_index('sample')
```

```python
df.head()
```

```python
# split df into array containing features
X = df.values
#X_df = df.drop(['sample'], axis=1)

ks = range(1, 16) # number of k
inertias = [] # list of intertias

# function to create elbow plot for determining optimal k using intertia
for k in ks:
  # create KMeans instance with k clusters: model
  model = KMeans(n_clusters=k)

  # fit model to samples
  model.fit(X)

  # append intertia value to list of intertias
  inertias.append(model.inertia_)

# plot ks versus intertias
plt.plot(ks, inertias, '-o')
plt.xlabel('number of clusters, k')
plt.ylabel('inertia')
plt.xticks(ks)
plt.savefig('kmeans_elbow_plot.pdf', bbox_inches='tight')
```
