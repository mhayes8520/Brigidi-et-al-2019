```python
import scipy.stats as stats
import numpy as np
import pandas as pd
from statsmodels.sandbox.stats.multicomp import multipletests as mt
```

```python
tss_motif_stuff = pd.read_csv("tssAnn_forFisherExact", sep = "\t")
tss_motif_stuff
```

```python
fet = lambda a,b,c,d:stats.fisher_exact([[a, b], [c, d]])
```

```python
fet_sArg = lambda arg:fet(*arg)
```

```python
list(map(fet_sArg, tss_motif_stuff[["Tss + in cluster", "tss not cluster", "cluster not tss", "neither"]].values))
```

```python
odds_ratios,pvals = zip(*list(map(fet_sArg, tss_motif_stuff[["Tss + in cluster", "tss not cluster", "cluster not tss", "neither"]].values)))
```

```python
tss_motif_stuff["odds ratios"] = odds_ratios
tss_motif_stuff["p values"] = pvals
tss_motif_stuff["-log(pval+1E-250)"] = -np.log10(tss_motif_stuff["p values"]+10**-250)
tss_motif_stuff
```

```python
tss_motif_stuff["p adj"] = mt(tss_motif_stuff["p values"], method = "fdr_bh")[1]
tss_motif_stuff["-log(p adj+1E-250)"] = -np.log10(tss_motif_stuff["p adj"]+10**-250)
tss_motif_stuff
```
